This project explores recreational and public amenities in Hyde Park, Chicago, using OpenStreetMap (OSM) data.  
The goal was to extract and visualize features such as gardens, swimming pools, parking, and bicycle rentals using SQL queries in QGIS.


#### Data and Tools

| Tool | Purpose |
|------|----------|
| QGIS 3.x | Spatial visualization, SQL queries, map layout |
| OSM (.osm) file | Geographic data source |
| QGIS Virtual Layers | SQL queries without PostGIS |
| Jupyter Notebook | Documentation and presentation |


#### Data Preparation

1. Imported `hyde_park.osm`
2. Created and examined the following layers:
   - `points`
   - `lines`
   - `multipolygons`
3. Verified available attributes:
   - `amenity`, `leisure`, `name`
4. Noted that certain attributes (like `cafe` or `park`) were missing in this dataset.


```python
-- Step 1: Recreation Features
SELECT *
FROM multipolygons
WHERE leisure IN ('garden', 'swimming_pool', 'outdoor_seating');
```

_This query selects recreational features such as gardens and swimming pools from the multipolygons layer._

```python
-- Step 2: Transportation/Utility Features
SELECT *
FROM multipolygons
WHERE amenity IN ('parking', 'shelter', 'bicycle_rental');
```

_This query extracts transport-related and public utility features (e.g., parking and bicycle rentals)._
